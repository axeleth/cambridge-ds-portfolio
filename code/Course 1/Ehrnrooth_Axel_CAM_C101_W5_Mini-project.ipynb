The data set contains six important features continuously monitored to evaluate the engine's status as ‘good’ or ‘bad’. These features are:
- **Engine rpm (revolutions per minute):** A high rpm indicates the engine is operating at a higher speed than designed for prolonged periods, which can lead to overheating, excessive wear, and eventual failure. A low rpm could signal a lack of power, issues with fuel delivery, or internal mechanical problems.

- **Coolant pressure:** Low coolant pressure indicates a potential leak in the cooling system or a coolant pump failure, risking engine overheating. A high coolant pressure could be a sign of a blockage in the cooling system or a failing head gasket, which can also lead to overheating.
- **Lubrication oil temperature:** High lubrication oil temperature suggests the oil is overheating, which can degrade its lubricating properties and lead to engine damage. A low lubrication oil temperature may indicate it is not reaching its optimal operating temperature, potentially causing inadequate lubrication.
- **Coolant temperature:** High coolant temperature signals overheating, which various issues, including a failed thermostat, coolant leak, or insufficient coolant flow can cause. A low coolant temperature could suggest the engine is not reaching its optimal operating temperature, affecting performance and efficiency.

Issues with engines could lead to engine malfunctions, potential safety hazards, and downtime (e.g. delayed deliveries), resulting in the breakdown of a ship’s overall functionality, consequently impacting the business, such as affecting revenue via failure to deliver goods. # **Performing Anomaly Detection**
**Now that the data has been inspected and visualised, it is time to analyse the outliers. In this section I will perform anomaly detection using various methods.**
#IQR Method for anomaly detection:

```python
data_w_outliers = engine_data.copy()

data_w_outliers.insert(1, 'Engine rpm outlier', 0)
data_w_outliers.insert(3, 'Lub oil pressure outlier', 0)
data_w_outliers.insert(5, 'Fuel pressure outlier', 0)
data_w_outliers.insert(7, 'Coolant pressure outlier', 0)
data_w_outliers.insert(9, 'lub oil temp outlier', 0)
data_w_outliers.insert(11, 'Coolant temp outlier', 0)
```

```python
# IQR method for indentifying outliers


Q1 = engine_data.quantile(0.25)
Q3 = engine_data.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify oultiers

outlier_mask = (engine_data < (Q1 - 1.5 * IQR)) | (engine_data > (Q3 + 1.5 * IQR))

for i in outlier_mask.columns:
  data_w_outliers[f'{i} outlier'] = outlier_mask[i].astype(int)
```

```python
data_w_outliers.head()
```

```python
for column in data_w_outliers.columns:
  if column.endswith('outlier'):
    print(data_w_outliers[column].value_counts())
```

```python
boolean_columns = [col for col in data_w_outliers.columns if col.endswith('outlier')]

filtered_df = data_w_outliers[data_w_outliers[boolean_columns].sum(axis=1) >= 2]

proportion_of_outliers = filtered_df.shape[0] / engine_data.shape[0]
```

```python
print(f"""When a data entry has at least two features seen as outliers,
the percentage of outliers in the data is {proportion_of_outliers * 100:.2f}%""")
```

#Anomaly detection with Machine Learning:

I will be scaling the data by standardization according to the Z-scores. This method was chosen because it is suitable for data that contains outliers.

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(engine_data)

# Turning ndarray into a data frame
engine_data_scaled = pd.DataFrame(scaled_data, columns=engine_data.columns)
engine_data_scaled.head()
```

****

# One-class Support Vector Machine

```python
# Perform PCA to reduce the dimensionality of the data
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
ED_scaled_pca = pca.fit_transform(engine_data_scaled)

# keeping the data as an ndarray for compatibility with plotting function
```

```python
# Creating SVM model
gamma = 0.1; nu = 0.012

model = OneClassSVM(kernel='rbf', nu=nu, gamma=gamma) # Using RBF kernel
model.fit(ED_scaled_pca) # train

print(f'SVM trained with gamma: {gamma}, and Nu: {nu}')

# Predicting anomalies using the trained SVM
y_pred = model.predict(ED_scaled_pca)
# print(len(y_pred))

# Adding an anomaly boolean column to the data frame
engine_data['OCSVM_Anomalies'] = y_pred
# engine_data.head()
```

```python
# Displaying the data entries that are classified as anomalies
anomalies = engine_data[engine_data['OCSVM_Anomalies'] == -1]
anomalies
```

```python
# I am utilising the plotting function provided in demonstration 5.2.3 on applying one-class SVM

# Visualising anomalies and the decision boundary
def plot_model(model, X, title):
  xx, yy = np.meshgrid(np.linspace(-5, 5, 500), np.linspace(-5, 5, 500))
  Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)


  #Predict whether each point is an inlier (1) or an outlier (-1)
  y_pred = model.predict(X)

  # Plot setup
  plt.figure(figsize=(12, 8))
  plt.title(title)
  plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuBu, alpha=0.8)
  plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='darkred') # decision boundary
  plt.contourf(xx, yy, Z, levels=[0, Z.max()], colors='orange', alpha=0.5) # Anomaly regions


  # Plot the anomalies
  sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=np.where(y_pred == 1, 'Normal', 'Anomaly'),
                  style=np.where(y_pred == -1, 'Anomaly', 'Normal'), markers={'Anomaly': 'X', 'Normal': 'o'},
                  palette = {'Normal': 'deepskyblue', 'Anomaly': 'red'}, alpha=0.6, edgecolor='k')
  # raise Exception(yy)

  # Plot aesthetics
  plt.axis('tight')
  plt.xlim((-5, 5))
  plt.ylim((-5, 5))
  plt.xlabel('Principal Component 1')
  plt.ylabel('Principal Component 2')
  plt.legend()
  plt.show()
```

```python
plot_model(model, ED_scaled_pca, "One-class SVM for Anomaly Detection")
```

```python
proportion_of_outliers_SVM = anomalies.shape[0] / engine_data.shape[0]
print(f"""Using a One-class SVM with gamma: {gamma}, and Nu: {nu},
the proportion of anomalies becomes {proportion_of_outliers_SVM * 100:.2f}%""")
```

# Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
# Turning the Scaled + PCA data into a data frame
ED_scaled_pca_df = pd.DataFrame(ED_scaled_pca, columns=['PC1', 'PC2'])

# will still be using the ndarray for the forest model
```

```python
data_2D = engine_data_scaled.copy()
data_2D['anomaly'] = y_pred
data_2D.head()
```

```python
# INIT model
n_estimators = 100; contamination = 0.03; random_state = 42

iso_forest = IsolationForest(n_estimators=n_estimators, contamination=contamination, random_state=random_state)
iso_forest.fit(ED_scaled_pca)
```

```python
# Predict scores

scores_pred = iso_forest.decision_function(ED_scaled_pca)

y_pred = iso_forest.predict(ED_scaled_pca)

# predict and count anomalies
ED_scaled_pca_df['anomaly'] = y_pred

ED_scaled_pca_df.head() # display
```

```python
# display anomalies only
anomalies = ED_scaled_pca_df[ED_scaled_pca_df.anomaly == -1]
anomalies
```

```python
# Visualising Forest

plt.figure(figsize=(15, 10))

plt.scatter(ED_scaled_pca_df.iloc[:, 0],
            ED_scaled_pca_df.iloc[:, 1],
            color='lightblue',
            s=50)


# Highlight anomalies.
plt.scatter(ED_scaled_pca_df[ED_scaled_pca_df.anomaly == -1].iloc[:, 0],
            ED_scaled_pca_df[ED_scaled_pca_df.anomaly == -1].iloc[:, 1],
            c='red', s=50, label='Anomalies')

# Add labels for anomalies.
for i in anomalies.index:
    plt.annotate(i, (ED_scaled_pca[i, 0], ED_scaled_pca[i, 1]),
                 textcoords="offset points",
                 xytext=(5, -5),
                 ha='left', va='center', fontsize=7)

# Add title, labels, and legend.
plt.title("Isolation Forest for Anomaly Detection", fontsize=20)
plt.xlabel('Principal Component 1', fontsize=10)
plt.ylabel('Principal component 2', fontsize=10)
plt.legend()

# View plot.
plt.show()
```

```python
# Proportion

proportion_of_outliers_forest = anomalies.shape[0] / engine_data.shape[0]
print(f"""Using an Isolation Forest with n_estimators: {n_estimators}, and contamination: {contamination},
the proportion of anomalies becomes {proportion_of_outliers_forest * 100:.2f}%""")
```
