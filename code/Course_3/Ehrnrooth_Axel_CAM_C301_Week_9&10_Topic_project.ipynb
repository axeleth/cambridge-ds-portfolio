```python
# Date range from first ever sale to last ever sale
full_range = pd.date_range(start=master.index.min(), end=master.index.max(), freq='W-SAT')
full_range.min(),full_range.max()
```

```python
# Advanced Re-Indexing

def reindex_advanced(group):
  """
  Made this function to avoid the Metadata like Title and Author from disappearing
  when resampling the traditional way. May seem long-winded but it worked out just
  fine.
  """
  reindexed = group.reindex(full_range) # ALL dates as indexes

  # Values that Stay Constant
  for col in ['ISBN', 'Title', 'Author', 'RRP', 'Binding', 'Imprint', 'Publisher Group', 'Product Class', 'Group', 'Publication Date', 'Country of Publication']:
    reindexed[col] = group[col].iloc[0]  # Same for all

  # Filled with 0's
  for col in ['Volume', 'Value', 'ASP']:
    reindexed[col] = reindexed[col].fillna(0)

  return reindexed
```

```python
resampled = master.groupby('ISBN', group_keys=False).apply(reindex_advanced)
```

```python
# Rename Index
resampled = resampled.reset_index().rename(columns={'index': 'End Date'})
```

## Decomposition


Inspecting the graphs above, both books have seasonal sales spikes before Christmas. Variation in the magnitude of the seasonal spikes can clearly be seen for both books. this variability makes Seasonal-Trend-Loess (STL) decomposition a sutiable method in both cases due to its ability to precisely model more complex variability in seasonal trends.


```python
# The Very Hungry Caterpillar STL

# Date as index for STL dataframe
stl_cat_df = caterpillar.set_index('End Date')
stl_cat_df = stl_cat_df.sort_index()

stl_cat = STL(stl_cat_df['Volume'], period=52)
res_cat = stl_cat.fit()

# plot decopmosition
res_cat.plot()
plt.tight_layout()
plt.show()
```

```python
# The Alchemist STL

# Date as index for STL dataframe
stl_alc_df = alchemist.set_index('End Date')
stl_alc_df = stl_alc_df.sort_index()

stl_alc = STL(stl_alc_df['Volume'], period=52)
res_alc = stl_alc.fit()

# plot decopmosition
res_alc.plot()
plt.tight_layout()
plt.show()
```

```python
# Autocorrelation for TVHC
acorr_cat = caterpillar['Volume']

fig, axes = plt.subplots(2,1,figsize=(8, 8))

# ACF
plot_acf(acorr_cat, lags=52, ax=axes[0])
axes[0].set_title('ACF - The Very Hungry Caterpillar')

# PACF
plot_pacf(acorr_cat, lags=52, ax=axes[1])
axes[1].set_title('PACF - The Very Hungry Caterpillar')

plt.tight_layout()
plt.show()
```

```python
# Autocorrelation for Alchemist
acorr_alc = alchemist['Volume']

fig, axes = plt.subplots(2,1,figsize=(8, 8))

# ACF
plot_acf(acorr_alc, lags=52, ax=axes[0])
axes[0].set_title('ACF - The Alchemist')

# PACF
plot_pacf(acorr_alc, lags=52, ax=axes[1])
axes[1].set_title('PACF - The Alchemist')

plt.tight_layout()
plt.show()
```

The ACF plot for *The Very Hungry Caterpillar* shows an initial quick decay followed by a gradual decline up to around lag 45, after which it briefly dips into the confidence zone and then rises again at lags near 52. This suggests a mix of short-term dependencies and strong yearly seasonality. The PACF plot reveals significant spikes at lags 1 and 2, indicating an AR(2) process.


The ACF for *The Alchemist* displays a sharp initial decline, becoming statistically insignificant after lag 10, with a clear resurgence at lag 52, indicating strong annual seasonality. The PACF plot shows significant spikes at lags 1 and 2, suggesting an AR(2) structure, and additional seasonal spikes near lag 52 further reinforce the presence of a seasonal AR component.

*The Very Hungry Caterpillar* shows a more sustained, long-memory sales pattern, while *The Alchemist* displays short-term, reactive behavior with a sharp decay in influence from past weeks.


## Stationarity
**Will conduct ADF-test to evaluate stationarity.**

```python
cat_adf = adfuller(caterpillar['Volume'])

print('ADF Statistic: %f' % cat_adf[0])
print('p-value: %f' % cat_adf[1])

if cat_adf[1] <= 0.05:
  print('The series is stationary')
else:
  print('The series is not stationary')
```

```python
alc_adf = adfuller(alchemist['Volume'])

print('ADF Statistic: %f' % alc_adf[0])
print('p-value: %f' % alc_adf[1])

if alc_adf[1] <= 0.05:
  print('The series is stationary')
else:
  print('The series is not stationary')
```

The ADF test shows that both time series are stationary. Notable, however, is the fact that *The Alchemist* is more stationary than *The Very Hungry Caterpillar*. This observation does make sense, as we saw that *The Very Hungry Caterpillar* has grown in popularity over time and therefore might carry a slightly stronger indication towards a unit root.

```python
def prepare_lagged_data(data, lag_window, forecast_horizon=32):

    df = data.copy()

    # Create lag features
    for i in range(1, lag_window + 1):
        df[f'lag_{i}'] = df['Volume'].shift(i)


    # Train/test split
    train_data = df.iloc[:-forecast_horizon]
    test_data = df.iloc[-forecast_horizon:]

    # Define features
    cols = [f'lag_{i}' for i in range(1, lag_window + 1)]

    # Extract final matrices
    X_train = train_data[cols].values
    y_train = train_data['Volume'].values

    X_test = test_data[cols].values
    y_test = test_data['Volume'].values

    return X_train, y_train, X_test, y_test
```

```python
def xgb_pipeline(data, horizon=32):

  ## INIT

  param_grid = {
      'n_estimators': [10, 50, 100],
      'max_depth': [i for i in range(1,15,3)]
  }

  lag_windows = [i for i in range(1,22,2)]  # values to try for window_length

  best_model = None
  best_mae = np.inf
  best_lag = None
  best_params = None

  ############
  ### TRAINING

  for lag in lag_windows:
      print(f"\nTrying window_size = {lag}")

      # prep data
      X_train, y_train, _, _ = prepare_lagged_data(data, lag, horizon)

      tscv = TimeSeriesSplit(n_splits=8) # Cross Validation with 8 expanding folds

      # Hyperparameter grid search
      xgb = XGBRegressor(objective='reg:squarederror', random_state=42)
      grid = GridSearchCV(xgb, param_grid, cv=tscv, scoring='neg_mean_absolute_error')


      grid.fit(X_train, y_train)

      # Evaluate model
      mean_mae = -grid.best_score_
      # print(f"MAE: {mean_mae:.2f} | Best params: {grid.best_params_}")

      # Track best model
      if mean_mae < best_mae:
          best_mae = mean_mae
          best_model = grid.best_estimator_
          best_lag = lag
          best_params = grid.best_params_


  ###########
  ### TESTING

  # Prepare test set
  X_train, y_train, X_test, y_test = prepare_lagged_data(data, best_lag, horizon)

  # Re-train on full training set
  best_model.fit(X_train, y_train)

  # Forecast
  forecast = best_model.predict(X_test)

  # Evaluate
  test_mae = mean_absolute_error(y_test, forecast)
  test_mape = mean_absolute_percentage_error(y_test, forecast)
  print(f"\nFinal Test (lag {best_lag}) MAE: {test_mae:.2f}, MAPE: {test_mape:.3f}")

  return best_model, forecast, test_mae, test_mape
```

```python
def perform_STL(df, period, index_sorted=True):

  # Sort index if not sorted
  if not index_sorted:
    df = df.sort_index()

  stl_df = STL(df['Volume'], period=period) # yearly
  res_df = stl_df.fit()

  # plot decopmosition
  res_df.plot()
  plt.tight_layout()
  plt.show()
```

### Autocorrelation

```python
def perform_AutoCorrelation(df, book, lags=12, index_sorted=True):

  # Sort index if not sorted
  if not index_sorted:
    df = df.sort_index()

  ##################
  # Autocorrelation
  acorr = df['Volume']

  fig, axes = plt.subplots(2,1,figsize=(8, 8))

  # ACF
  plot_acf(acorr, lags=lags, ax=axes[0])
  axes[0].set_title(f'ACF - {book}')

  # PACF
  plot_pacf(acorr, lags=lags, ax=axes[1])
  axes[1].set_title(f'PACF - {book}')

  plt.tight_layout()
  plt.show()
```

### Stationarity

```python
def perform_ADF(df, book, index_sorted=True):

  print(f"Stationarity of {book}")

  # Sort index if not sorted
  if not index_sorted:
    df = df.sort_index()

  df_adf = adfuller(df['Volume'])

  print('ADF Statistic: %f' % df_adf[0])
  print('p-value: %f' % df_adf[1])

  if df_adf[1] <= 0.05:
    print('The series is stationary')
  else:
    print('The series is not stationary')
```

### Performing Monthly Auto ARIMA

```python
# Data split
def forecast_horizon_split_m(df, horizon=8):
  train = df.iloc[:-horizon]
  test = df.iloc[-horizon:]
  return train, test

  # cat_train, cat_test = forecast_horizon_split(caterpillar, HORIZON)
  # alc_train, alc_test = forecast_horizon_split(alchemist, HORIZON)
```

```python
def perform_AutoARIMA(df, book, index_sorted=True):

  # Sort index if not sorted
  if not index_sorted:
    df = df.sort_index()

  train, test = forecast_horizon_split_m(df)



  model = auto_arima(train['Volume'],
                   start_p=0, max_p=5,
                   start_q=0, max_q=5,
                   start_P=0, max_P=3,
                   start_Q=0, max_Q=3,
                   d=None,
                   D=None,
                   seasonal=True,
                   m=12,
                   stepwise=True,         # Still efficient
                   trace=True,
                   suppress_warnings=True,
                   error_action='ignore', # avoid crashing on bad fits
                   approximation=False)   # ensure full search (not shortcut)


  return model, train, test
```
