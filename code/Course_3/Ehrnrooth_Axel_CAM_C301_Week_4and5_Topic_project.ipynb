##### Preprocessing

```python
# Preprocessing funciton
def preprocess_text(text, lang_code):
    """
    This funciton will preprocess the text by tokenizing, removing stopwords, and stemming the words.
    """
    lang_map = {
        'en': 'english',
        'de': 'german',
        'fr': 'french',
        'da': 'danish'
    }

    lang = lang_map[lang_code] # get the NLTK language naming convention

    tokens = word_tokenize(text, language=lang)
    tokens = [token.lower() for token in tokens]
    stop_words = set(stopwords.words(lang))
    filtered = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]

    return filtered
```

```python
# Function to apply language model with specific prompt to the reviews.

def extract_topics(review):
    prompt = f"""In the following customer review, pick out the main 3 topics. Return them in a numbered list format, with each one on a new line.\n\n{review}"""
    try:
        response = lang_model(prompt, batch_size=16)
        return response[0]['generated_text']
    except Exception as e:
        print(f"Error with review: {e}")
        return ""
```

```python
# Generate topics
topic_outputs = subset_llm["Review"].apply(extract_topics)
```

```python
# Cleaning the responses to keep the lists
def extract_clean_topics(text):
    return [re.sub(r"^\d+\.\s*", "", line.strip()) for line in text.split("\n") if re.match(r"^\d+\.", line.strip())]
```
