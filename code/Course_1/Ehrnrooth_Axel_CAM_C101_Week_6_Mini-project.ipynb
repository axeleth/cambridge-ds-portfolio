Upon inspection, it does NOT seem that the data we are interested in identifying outliers for is normally distributed.

Because the data is not normally distributed, **any hypothesis testing and anomaly detection methods used cannot assume a normal distribution.**

The IQR method is an appropriate way of determining outliers in this case since it does require the assumption of a normal distribution.
***

```python
features = ['Quantity', 'Total Revenue', 'Average Unit Costs', 'Profit', 'Customer Age']

# a data frame with the features from df_agg_CID described in the 'features' list
df_IQR_test = df_agg_CID[features]


for feature in features_and_bins.keys():
    Q1 = df_IQR_test[feature].quantile(0.25)
    Q3 = df_IQR_test[feature].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_IQR_test[(df_IQR_test[feature] < lower_bound) | (df_IQR_test[feature] > upper_bound)]
    print(f"Number of outliers in {feature}: {len(outliers)}")
```

**Customer Age** and **Average Unit Costs** have already been added to the data frame for earlier analysis.

The second feature that can be created is **Frequency**. This feature will give insight into how often a specific customer makes purchases.

```python
def average_time_between_orders(dates):
  """
  Average time between order dates:

  1. Remove duplicate dates
  2. If there are more than 2 dates:
        calculate time differences between consecutive dates
  3. Return average in days

  NOTE: The higher the frequency value, the less frequently a customer shops
  """

  unique_dates = sorted(set(dates))

  if len(unique_dates) > 1:
    deltas = np.diff(unique_dates)
    return np.mean(deltas).days

  return None
```

```python
df_agg_CID['Frequency'] = df_agg_CID['Order_Date'].apply(average_time_between_orders)
df_agg_CID.head()
```

***
Another interesting feature worth adding is **Recency**. Looking at how recently a customer has made their last purchase can give great insight into churn rates over time.

This feature will be stored as the number of days since the last order date.

```python
def recency(dates):
  """
  From a list of dates, return the amount of days since the most recent date.

  1. Find the most recent date
  2. Calculate the amount of days since that date
  3. Return result
  """
  most_recent = max(dates) # most recent date

  return (pd.to_datetime('today') - most_recent).days
```

```python
df_agg_CID['Recency'] = df_agg_CID['Order_Date'].apply(recency)
df_agg_CID.head()
```

***
The next feature to be engineered is **Customer Lifetime Value (CLV)**. This is a feature that combines multiple key factors of one customer to evaluate how much value they are still likely to add to the company. It takes Age, revenue, frequency, and recency into account to create an index for how valuable a customer is to the company.

```python
def calculate_clv(data):
  """
  data = an entry in the data

  Formula: CLV = (Average Order Revenue x (1/Frequency) x Estimated Customer Lifespan)
  """
  # Average Revenue vv
  dates = data['Order_Date']
  unique_dates = sorted(set(dates))
  avg_rev = data['Total Revenue']/len(unique_dates)

  # Frequency vv
  if pd.isna(data['Frequency']):
    frequency = float('inf') # dividing by infinity to avoid division by 0
  else:
    frequency = data['Frequency']

  # Estimated Customer Lifespan vv
  base_lifespan = 365 * 5
  customer_age_factor = max(0.1, 1 - (data['Customer Age'] / 100))  # older customers have shorter customer lifespan
  mean_recency = df_agg_CID['Recency'].mean()
  recency_factor = max(0.1, 1 - (data['Recency'] / mean_recency))  # Recency < mean recency = boosts lifespan

  estimated_lifespan = base_lifespan * customer_age_factor * recency_factor

  clv = avg_rev * (1/frequency) * estimated_lifespan

  return clv

```

```python
df_agg_CID['CLV'] = df_agg_CID.apply(calculate_clv, axis=1)
df_agg_CID.head()
```

```python
scaler = StandardScaler()
df_cl_sc = scaler.fit_transform(df_cluster.drop(columns=['Customer ID'])) # Scaled
df_cl_sc = pd.DataFrame(df_cl_sc, columns=df_cluster.columns[1:]) # Convert to dataframe
```

***
Using IQR method to identify EXTREME outliers in the data. Setting the upper and lower bounds to 10x Q1 and Q3 in order to omit as few entries as possible.
***

```python
# Identifying all outliers using IQR
Q1 = df_cl_sc.quantile(0.25)
Q3 = df_cl_sc.quantile(0.75)
IQR = Q3 - Q1

df_cl_no_outliers = df_cl_sc.copy()
df_cl_no_outliers['Outlier'] = 0  # Initializing an outlier column with 0
df_cluster['Outlier'] = 0 

for column in df_cl_sc.columns:
    Q1 = df_cl_sc[column].quantile(0.25)
    Q3 = df_cl_sc[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 10 * IQR
    upper_bound = Q3 + 10 * IQR
    outliers = (df_cl_no_outliers[column] < lower_bound) | (df_cl_no_outliers[column] > upper_bound)    
    df_cl_no_outliers.loc[outliers, 'Outlier'] = 1 # add a 1 to the 'Outlier' column if the row is an outlier
    df_cluster.loc[outliers, 'Outlier'] = 1 # add a 1 to the 'Outlier' column if the row is an outlier
```

***
Using the **Elbow** Method to determine the optimal value for k. This method will determine when an increase in k will no longer be beneficial
***

```python
wcss = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cl_no_outliers.drop(columns=['Outlier']))
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(axis='x', linestyle='--')
plt.show()
```

***
According to the Elbow method, the optimal value for k is 4 since this is where the within-cluster sum of squares (WCSS) visually starts to slow its decrease.

The second approach to determining the optimal k is the **Silhouette Scores** method. This method will determine how well the data points cluster.
***

```python
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cl_no_outliers.drop(columns=['Outlier']))
    silhouette_scores.append(silhouette_score(df_cl_no_outliers.drop(columns=['Outlier']), kmeans.labels_))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Scores')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(axis='x', linestyle='--')
plt.show()
```

***
Since both approaches showed optimal values for k=4, the obvious choice is to use the value 4 for k.

***

For the next approach, we perform hierarchical clustering to analyse the results reteived from the previous two methods. A hierarchical clustering structure will give another 'ballpark' confirmation for whether the suggestions for the optimal k are valid. Furthermore, it will work as a visual tool in the form of a dendrogram that will give insight into how and where clusters are divided.
***

```python
linked = linkage(df_cl_no_outliers.drop(columns=['Outlier']), method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked, no_labels=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.grid(axis='y', linestyle='--')
plt.show()

```

***
As seen in the dendrogram, the data is segmented into 3 clusters by the hierarchical clustering method. However, since the other two methods both consistently agreed on k=4 being optimal, and since k=3 is not an extreme difference from k=4, the former will be used. In terms of business context, k=4 gives the possibility of identifying a group of customers that do need attention that the hierarchical sturcture missed. 
***
